# Model Building

```python
# Splitting the columns
y = df_upsampled['Income']
x = df_upsampled.drop(['Income'],axis = 1)
```

```python
# Scaling Data using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
# Splitting the data for training and testing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier

maxaccu =0
maxrs = 0
for i in range(1,200):
    xtrain,xtest,ytrain,ytest=train_test_split(x_scaled,y,test_size=0.25,random_state=i)
    mod = DecisionTreeClassifier()
    mod.fit(xtrain,ytrain)
    pred = mod.predict(xtest)
    acc = accuracy_score(ytest,pred)
    if acc>maxaccu:
        maxaccu=acc
        maxrs =i
print("Best accuracy is ",maxaccu,"on random state ",maxrs)
```

After splitting the data best accuracy is 92% on random state is 160.

```python
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=.25, random_state=maxrs)
```

```python
#Importing Libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
```

DecisionTreeClassifier:

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)
acc = classification_report(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

SVC:

```python
sv = SVC()
sv.fit(X_train,y_train)
y_pred = sv.predict(X_test)
acc = classification_report(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

KNeighborsClassifier:

```python
knc = KNeighborsClassifier(n_neighbors=3)
knc.fit(X_train,y_train)
y_pred = knc.predict(X_test)
acc = classification_report(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

LogisticRegression:

```python
lr= LogisticRegression(random_state=1)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
acc = classification_report(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

AdaBoostClassifier:

```python
abc = AdaBoostClassifier(random_state=1)
abc.fit(X_train, y_train)
y_pred = abc.predict(X_test)
acc = classification_report(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

RandomForestClassifier:

```python
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
acc = classification_report(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

cross_val_score:

```python
from sklearn.model_selection import cross_val_score
```

```python
print(cross_val_score(dtc,x_scaled,y,cv=5).mean())
print(cross_val_score(rfc,x_scaled,y,cv=5).mean())
print(cross_val_score(lr,x_scaled,y,cv=5).mean())
print(cross_val_score(abc,x_scaled,y,cv=5).mean())
print(cross_val_score(knc,x_scaled,y,cv=5).mean())
print(cross_val_score(sv,x_scaled,y,cv=5).mean())
```

As per the above data of model building we conclued that RandomForestClassifier is the best model with the accuracy of 93%.

# Using Metrics:

```python
from sklearn import metrics

models =[dtc,rfc,sv,knc,lr,abc]
for i in models:
    metrics.plot_roc_curve(i, X_test, y_test)  
    plt.show()
```

As we see that Random Forest Classifier is the best model according to the metrics also.

# Hyper parameter Tuning

```python
GCV.best_params_
```

```python
Final_mod = RandomForestClassifier(criterion='gini',max_depth=19,max_features='sqrt')
Final_mod.fit(X_train,y_train)
pred = Final_mod.predict(X_test)
acc = accuracy_score(y_test,pred)
print(acc*100)
```

```python
confusion_matrix(y_test,pred)
```

```python
metrics.plot_roc_curve(Final_mod, X_test, y_test)  
plt.show()
```

Final Model using metrics gives 90% accuracy.
